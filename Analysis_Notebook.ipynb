## Hypothesis Testing

We have a new dataset that contains data on rides from the Loop to O'Hare International Airport. The dataset includes the following fields:

- `start_ts`: pickup date and time
- `weather_conditions`: weather conditions at the moment the ride started
- `duration_seconds`: ride duration in seconds

We are tasked with testing the following hypothesis:

"The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

Before we proceed, let's first examine the data in our new dataframe.

```python
# Display the first few rows of df3
df3.head()
```

## Data Types and Information

Next, we'll check the data types of the columns in our dataframe to ensure they are appropriate for our analysis. We'll also get some basic information about the dataframe.

```python
# Information about df3
df3.info()
```

## Data Types and Information - Summary

From the information above, we can see that in `df3`, we have 1068 entries. All columns (`start_ts`, `weather_conditions`, and `duration_seconds`) have no null values. The data types are as expected: `start_ts` is of type object (string), `weather_conditions` is of type object (string), and `duration_seconds` is a float.

However, for our analysis, it would be more appropriate if `start_ts` was in datetime format. Let's convert `start_ts` to datetime format and extract the day of the week for each entry.

```python
# Convert 'start_ts' to datetime format
df3['start_ts'] = pd.to_datetime(df3['start_ts'])

# Extract the day of the week and add it as a new column 'day_of_week'
df3['day_of_week'] = df3['start_ts'].dt.day_name()

# Display the first few rows of df3
df3.head()
```

```python
# find unique value in weather_conditions column
df3['weather_conditions'].unique()
```

## Formulating the Hypotheses

We are now ready to formulate our null and alternative hypotheses for the test. We are testing the hypothesis that "The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."

We can formulate our hypotheses as follows:

- Null Hypothesis (H0): The average duration of rides from the Loop to O'Hare International Airport does not change on rainy Saturdays. In other words, the average duration of rides is the same on rainy Saturdays as on other days.

- Alternative Hypothesis (H1): The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays. In other words, the average duration of rides is different on rainy Saturdays compared to other days.

We will use a significance level (alpha) of 0.05 for our test. This means that we are willing to accept a 5% chance of rejecting the null hypothesis when it is true.

## Preparing the Data for Hypothesis Testing

Before we can perform our hypothesis test, we need to prepare our data. We'll create two separate dataframes: one for rides on rainy Saturdays and one for rides on other days. These dataframes will be used to compare the average ride durations.

```python
# Create a dataframe for rides on rainy Saturdays
rainy_saturdays = df3[(df3['weather_conditions'] == 'Bad') & (df3['day_of_week'] == 'Saturday')]

# Create a dataframe for rides on other days
other_days = df3[(df3['weather_conditions'] == 'Good') | (df3['day_of_week'] != 'Saturday')]
```

## Hypothesis Testing

Now that we have our data prepared, we can proceed with the hypothesis testing. We'll use the t-test for independent samples, as we are comparing the means of two independent groups (rides on rainy Saturdays vs rides on other days).

The t-test assumes that the data is normally distributed and that the variances of the two groups are equal. However, even if these assumptions are not fully met, the t-test can be quite robust for large sample sizes.

```python
from scipy import stats

# Perform t-test
t_stat, p_value = stats.ttest_ind(rainy_saturdays['duration_seconds'], other_days['duration_seconds'], equal_var=False)

# Print the p-value
p_value
```

## Hypothesis Test Results

The p-value of our t-test is extremely small (6.74e-12), which is much less than our significance level of 0.05. Therefore, we reject the null hypothesis that the average duration of rides from the Loop to O'Hare International Airport does not change on rainy Saturdays.

This suggests that the average duration of rides from the Loop to O'Hare International Airport does indeed change on rainy Saturdays compared to other days. However, this does not necessarily mean that the change is due to the weather conditions or the day of the week. Other factors could also be influencing the ride durations.
